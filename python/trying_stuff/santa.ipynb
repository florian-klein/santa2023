```python
def get_generator_set_from_puzzle_type(puzzle_type):
    """
    Returns list of generators (Permutation as specified in csv for allowed moves)
    """
    generator_set = []
    for _, move_permutation in allowed_moves[puzzle_type].items():
        generator_set.append(Permutation(move_permutation))
    return generator_set

generator_set = get_generator_set_from_puzzle_type('cube_2/2/2')
```

```python
permutation_group = PermutationGroup(generator_set)
from functools import reduce
def get_generator_decomp(g, permutation_to_move_name, permutation_group):
    """
    Uses permutation_group to decompose g into a product of generators
    """
    permutation_decomp = permutation_group.generator_product(g, original=True)
    move_names = []
    for permutation in permutation_decomp:
        move_names.append(permutation_to_move_name[tuple(permutation)])
    return (len(move_names), ".".join(move_names))
```

```python
# ! WARN: This function only works if each element in the string is unique AND wildcards == 0!!
def find_unique_permutation(start_string, goal_string):
    """
    Returns the permutation that transforms the start state to the goal state
    WARN: This function only works if each element in the string is unique AND wildcards == 0!!
    """
    goal_element_to_index = {}
    for i, element in enumerate(goal_string):
        goal_element_to_index[element] = i
    permutation = []
    for element in start_string:
        permutation.append(goal_element_to_index[element])
    return permutation

def find_permutation_for_id(id):
    """
    Returns the permutation that transforms the start state to the goal state
    WARN: This function only works if each element in the string is unique AND wildcards == 0!!
    """
    operation_list_start = id_to_puzzle["initial_state"][id].split(";")
    operation_list_goal = id_to_puzzle["solution_state"][id].split(";")
    return find_unique_permutation(operation_list_start, operation_list_goal)

goal_perm_list = find_permutation_for_id(369) # this is the g we want the decomposition for
print(goal_perm_list)
# -d0.-d0.-r0.d0.r0.d0.d0.r1.f0.r0.d0.f1.r0.-d0.-d0.-r0.d0.r0.d0.d0.r0.-d0.-r0.d0.r0.-d0.-d0.-r0.d0.r0.d0.d0.d0.-d0.-d0.-r0.d0.r0.d0.d0.-d0.-r0.d0.r0
```
