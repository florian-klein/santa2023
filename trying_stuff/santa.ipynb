```python
allowed_moves = {}

for _, row in type_to_moves.iterrows():
    allowed_moves[row['puzzle_type']] = eval(row['allowed_moves'])
print(allowed_moves["cube_10/10/10"])
```

```python


def get_permutation_to_name_from_puzzle_type(puzzle_type):
    """
    Returns dict: permutation (tuple) -> move_name (str) as specified in puzzle_info.csv
    """
    permutation_to_move_name = {}

    for move_name, move_permutation in allowed_moves[puzzle_type].items():
        perm = Permutation(move_permutation)
        perm_inv = perm**-1
        permutation_to_move_name[tuple(perm)] = move_name
        permutation_to_move_name[tuple(perm_inv)] = "-" + move_name 
    return permutation_to_move_name

def get_name_to_permutation_from_puzzle_type(puzzle_type):
    """
    Returns dict: move_name (str) -> permutation (tuple) as specified in puzzle_info.csv
    """
    name_to_permutation = {}

    for move_name, move_permutation in allowed_moves[puzzle_type].items():
        perm = Permutation(move_permutation)
        perm_inv = perm**-1
        name_to_permutation[move_name] = tuple(perm)
        name_to_permutation["-" + move_name] = tuple(perm_inv)
    return name_to_permutation

permutation_to_move_name = get_permutation_to_name_from_puzzle_type('cube_2/2/2')
name_to_permutation = get_name_to_permutation_from_puzzle_type('cube_33/33/33')
```

```python
def get_generator_set_from_puzzle_type(puzzle_type):
    """
    Returns list of generators (Permutation as specified in csv for allowed moves)
    """
    generator_set = []
    for _, move_permutation in allowed_moves[puzzle_type].items():
        generator_set.append(Permutation(move_permutation))
    return generator_set

generator_set = get_generator_set_from_puzzle_type('cube_2/2/2')
```

```python
permutation_group = PermutationGroup(generator_set)
from functools import reduce
def get_generator_decomp(g, permutation_to_move_name, permutation_group):
    """
    Uses permutation_group to decompose g into a product of generators
    """
    permutation_decomp = permutation_group.generator_product(g, original=True)
    move_names = []
    for permutation in permutation_decomp:
        move_names.append(permutation_to_move_name[tuple(permutation)])
    return (len(move_names), ".".join(move_names))
```

```python
# ! WARN: This function only works if each element in the string is unique AND wildcards == 0!!
def find_unique_permutation(start_string, goal_string):
    """
    Returns the permutation that transforms the start state to the goal state
    WARN: This function only works if each element in the string is unique AND wildcards == 0!!
    """
    goal_element_to_index = {}
    for i, element in enumerate(goal_string):
        goal_element_to_index[element] = i
    permutation = []
    for element in start_string:
        permutation.append(goal_element_to_index[element])
    return permutation

def find_permutation_for_id(id):
    """
    Returns the permutation that transforms the start state to the goal state
    WARN: This function only works if each element in the string is unique AND wildcards == 0!!
    """
    operation_list_start = id_to_puzzle["initial_state"][id].split(";")
    operation_list_goal = id_to_puzzle["solution_state"][id].split(";")
    return find_unique_permutation(operation_list_goal, operation_list_start)

goal_perm_list = find_permutation_for_id(21) # this is the g we want the decomposition for
print(goal_perm_list)
goal_permutation = Permutation(goal_perm_list)
length, move_decomp = get_generator_decomp(goal_permutation, permutation_to_move_name, permutation_group)
print("Length of decomposition: ", length)
print("Move decomposition: ", move_decomp)
# -d0.-d0.-r0.d0.r0.d0.d0.r1.f0.r0.d0.f1.r0.-d0.-d0.-r0.d0.r0.d0.d0.r0.-d0.-r0.d0.r0.-d0.-d0.-r0.d0.r0.d0.d0.d0.-d0.-d0.-r0.d0.r0.d0.d0.-d0.-r0.d0.r0
```

```python
def parse_move_string_to_permutation(move_string, name_to_permutation):
    move_list = move_string.split(".")
    move_list = list(map(lambda x: Permutation(name_to_permutation[x]), move_list))
    state = Permutation([])
    for move in move_list:
        state = move * state
    return state

def assert_permutation_equal(p1_str, p2_str, name_to_permutation):
    p1 = parse_move_string_to_permutation(p1_str, name_to_permutation)
    p2 = parse_move_string_to_permutation(p2_str, name_to_permutation)
    if p1 == p2:
        print("Permutations are equal")
    else:
        print("Permutations are not equal")
        print(p1)
        print(p2)

print(list(parse_move_string_to_permutation(submission["moves"][21], name_to_permutation)))
print(list(parse_move_string_to_permutation(move_decomp, name_to_permutation)))
assert_permutation_equal(move_decomp, submission["moves"][21], name_to_permutation)
```

```python
def constrained_sort(unsorted_string, cycle_permutations):
    graph = build_graph(unsorted_string, cycle_permutations)
    connected_components = find_connected_components(graph)

    sorted_string = ""
    for component in connected_components:
        sorted_component = sort_within_component(component)
        sorted_string += sorted_component

    return sorted_string

# Additional helper functions

def build_graph(unsorted_string, cycle_permutations):
    graph = {}  # Using an adjacency list to represent the graph

    # Create nodes for each character in the string
    for char in unsorted_string:
        graph[char] = []

    # Add edges based on the given cycle permutations
    for cycle in cycle_permutations:
        add_edges(graph, cycle)

    return graph

def add_edges(graph, cycle):
    # Add edges for a cycle permutation
    for i in range(len(cycle)):
        graph[cycle[i]].append(cycle[(i + 1) % len(cycle)])
        graph[cycle[(i + 1) % len(cycle)]].append(cycle[i])

def find_connected_components(graph):
    # Use depth-first search (DFS) to find connected components
    components = []
    visited = set()

    def dfs(node, component):
        visited.add(node)
        component.append(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor, component)

    for node in graph:
        if node not in visited:
            component = []
            dfs(node, component)
            components.append(component)

    return components

def sort_within_component(component):
    # Sort characters within a connected component
    return ''.join(sorted(component))

# Example Usage
unsorted_string = "bca"
cycle_permutations = [("a", "b", "c")]
result = constrained_sort(unsorted_string, cycle_permutations)
print(result)
```
