```python
import pandas as pd
from sympy.combinatorics import Permutation
```

```python
targets = pd.read_csv("./target.csv")
puzzles = pd.read_csv("./puzzles.csv")
puzzles_info = pd.read_csv("./puzzle_info.csv")
puzzles_info.head()
```

```python
def get_puzzle_type_for_id(puzzle_id):
    return puzzles[puzzles["id"] == puzzle_id].values[0][1]

def get_perms_for_id(puzzle_id):
    perms = []
    names = []
    puzzle_type = get_puzzle_type_for_id(puzzle_id)
    allowed_moves = eval(puzzles_info[puzzles_info["puzzle_type"] == puzzle_type].values[0][1])
    for (name, moves) in allowed_moves.items():
        perms.append(Permutation(moves))
        names.append(name)
    return perms, names

perms, name = get_perms_for_id(20)

def increment_nums_in_str(s):
    return " ".join(list(map(lambda x: str(int(x) + 1), s.split(" "))))
def get_string_version_of_perm(perm):  
    return "".join(list(filter(lambda x: len(x.split(",")) != 1, map(lambda x: "(" + increment_nums_in_str(x.replace("(", "").replace(")", "")).replace(" ", ",") + ")", str(perm).split(")(")))))

def get_gap_gen_group(id):
    perms, names = get_perms_for_id(id)
    perms = ",".join(list(map(lambda x: get_string_version_of_perm(x), perms)))
    res_string = "g:=Group(" + perms + ");\n"
    homs = ",".join(list(map(lambda x: "\""+str(x)+"\"", names)))
    res_string += 'hom:=EpimorphismFromFreeGroup(g:names:=['+homs+']);\n'
    target = get_string_version_of_perm(Permutation(list(map(lambda x: x - 1, eval("[" + targets[targets["id"] == id].values[0][1] + "]")))))
    res_string += "word := PreImagesRepresentative( hom, "+target+" );\n"
    res_string += "Length( word );"
    return res_string

print(get_gap_gen_group(283))

```
